# Randomized Search Cross-Validation (RandomizedSearchCV) Model
 Using Randomized Search Cross -Validation Model



*   Pro: It is computationally more efficient and faster than Grid Search as it does not evaluate every possible combination of parameters, making it suitable for large datasets and wide parameter ranges.
* Con: It may miss the optimal hyperparameter combination because it randomly samples from the parameter space, which means not all parameter values are tested.


```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_dist_dt = {
    'max_depth': [None] + list(np.arange(3, 15)),
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 10),
    'criterion': ['gini', 'entropy'],
    'max_features': ['auto', 'sqrt', 'log2']
}

random_search_dt = RandomizedSearchCV(
    estimator=DecisionTreeClassifier(random_state=42),
    param_distributions=param_dist_dt,
    n_iter=100,
    cv=5,
    random_state=42,
    n_jobs=-1
)

random_search_dt.fit(trainInput, trainTarget)

best_decision_tree = random_search_dt.best_estimator_

best_decision_tree_predictions = best_decision_tree.predict(validationInput)


best_decision_tree_accuracy = accuracy_score(validationTarget, best_decision_tree_predictions)
print("Best Decision Tree Accuracy:", best_decision_tree_accuracy)
print(classification_report(validationTarget, best_decision_tree_predictions))
```

# Bayesian Optimization (with libraries like scikit-optimize)
Using Bayesian Optimization Model


Bayesian Optimization uses probability to model the function that maps hyperparameters to their objective scores, and it iteratively updates this model to focus on finding the optimal values.



*   Pros: Efficient in finding the best hyperparameters with fewer evaluations.

* Cons: Can be more complex to set up compared to other techniques.







```python
!pip install scikit-optimize

from skopt import BayesSearchCV
from skopt.space import Real, Categorical, Integer

param_space = {
    'C': Real(1e-6, 1e+6, prior='log-uniform'),
    'kernel': Categorical(['linear', 'rbf', 'poly']),
    'gamma': Real(1e-6, 1e+1, prior='log-uniform'),
    'degree': Integer(1, 5),
}


svc_model = SVC(random_state=42)

opt = BayesSearchCV(
    svc_model,
    param_space,
    n_iter=32,
    cv=5,
    n_jobs=-1,
    random_state=42
)

opt.fit(trainInput, trainTarget)


best_svc_model = opt.best_estimator_

best_svc_predictions = best_svc_model.predict(validationInput)

best_svc_accuracy = accuracy_score(validationTarget, best_svc_predictions)
print("Best SVC Accuracy:", best_svc_accuracy)
print(classification_report(validationTarget, best_svc_predictions))
```

#  Naive Approach K Neighbours Classifier Model

```python
import numpy as np

def euclidean_distance(x1, x2):
  """Calculates the Euclidean distance between two points."""
  return np.sqrt(np.sum((x1 - x2) ** 2))

def knn_predict(X_train, y_train, X_new, k):
  """Predicts the class label for a new data point using the k-nearest neighbors algorithm."""
  y_pred = []
  for x_new in X_new:
    distances = [euclidean_distance(x_new, x_train) for x_train in X_train]
    k_nearest_indices = np.argsort(distances)[:k]
    k_nearest_labels = [y_train[i] for i in k_nearest_indices]

    pred_label = max(set(k_nearest_labels), key=k_nearest_labels.count)
    y_pred.append(pred_label)

  return np.array(y_pred)
```

# Hyperparameter Tuning with Particle Swarm Optimization (PSO)

Using the PSO Model.





PSO is inspired by the social behavior of birds. It uses particles that move through the parameter space to find optimal hyperparameters.

*   Pros: Efficient for multi-dimensional optimization.
*   Cons: Can be difficult to fine-tune the PSO parameters.

```python
!pip install pyswarms

import pyswarms as ps
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

def objective_function(x, X_train, y_train):
    """Objective function to minimize (1 - accuracy) for PSO."""
    costs = []
    for particle_position in x:
        n_neighbors = int(particle_position[0])
        weights = 'uniform' if particle_position[1] < 0.5 else 'distance'
        algorithm = KNeighborsClassifier(n_neighbors=n_neighbors, weights=weights)
        scores = cross_val_score(algorithm, X_train, y_train, cv=5)
        costs.append(1 - scores.mean())
    return np.array(costs)

lower_bound = [1, 0]
upper_bound = [15, 1]
bounds = (lower_bound, upper_bound)

options = {'c1': 0.5, 'c2': 0.3, 'w': 0.9}
optimizer = ps.single.GlobalBestPSO(n_particles=10, dimensions=2, options=options, bounds=bounds)

cost, pos = optimizer.optimize(objective_function, iters=100, X_train=trainInput, y_train=trainTarget)

best_n_neighbors = int(pos[0])
best_weights = 'uniform' if pos[1] < 0.5 else 'distance'

best_knn_model = KNeighborsClassifier(n_neighbors=best_n_neighbors, weights=best_weights)
best_knn_model.fit(trainInput, trainTarget)

best_knn_predictions = best_knn_model.predict(validationInput)

from sklearn.metrics import accuracy_score, classification_report
best_knn_accuracy = accuracy_score(validationTarget, best_knn_predictions)

print("Best KNN Accuracy with PSO:", best_knn_accuracy)
print(classification_report(validationTarget, best_knn_predictions))
```
